```python
import pandas as pd

# Masukkan dataset
df = pd.read_csv('hr_employee_attrition.csv')
```

```python
# Preview dataset
df.head()
```

```python
# Memeriksa tipe data 
df.info()
```

```python
# Mengecek total missing values
df.isnull().sum()
```

```python
# Mengecek unique values
df.nunique()
```

```python
# Menghilangkan kolom yang tidak dipakai (hanya punya 1 unique value)
df.drop(['EmployeeCount', 'Over18', 'StandardHours'], axis=1, inplace=True)
```

```python
df.info()
```

```python
# Mengubah nama kolom agar mudah dipahami
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
## Membuat kolom age_group
# Menentukan batas potongan umur
bins = [0, 24, 34, 44, 100]

# Menentukan labels
labels = ['Young', 'Early Pro', 'Mid-Career', 'Senior']

# Eksekusi cutting
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
# Preview hasil cutting
df[['age', 'age_group']].head(10)
```

```python
# Mengkategorikan kolom lama bekerja
bins_tenure = [-1, 2, 5, 10, 100]
labels_tenure = ['New Hire', 'Established', 'Loyal', 'Veteran']
df['tenure_status'] = pd.cut(df['yearsatcompany'], bins=bins_tenure, labels=labels_tenure)
```

```python
# Preview hasil cutting lama bekerja
df[['yearsatcompany', 'tenure_status']].head(10)
```

```python
## Decoding
# Mapping Education
edu_map = {1: 'Below College', 2: 'College', 3: 'Bachelor', 4: 'Master', 5: 'Doctor'}
df['education_label'] = df['education'].map(edu_map)
```

```python
df[['education', 'education_label']].head(10)
```

```python
# Mapping WorkLife Balance
wlb_map = {1: 'Bad', 2: 'Good', 3: 'Better', 4: 'Best'}
df['worklifebalance_label'] = df['worklifebalance'].map(wlb_map)
```

```python
df[['worklifebalance', 'worklifebalance_label']].head(10)
```

```python
# Mapping Performance rating
pr_map = {1: 'Low', 2: 'Good', 3: 'Excellent', 4: 'Outstanding'}
df['performance_label'] = df['performancerating'].map(pr_map)
```

```python
df[['performancerating', 'performance_label']].head(10)
```

```python
# Mapping Satisfaction & Involvement
sat_map = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
df['jobsat_label'] = df['jobsatisfaction'].map(sat_map)
df['envsat_label'] = df['environmentsatisfaction'].map(sat_map)
df['relsat_label'] = df['relationshipsatisfaction'].map(sat_map)
df['jobinv_label'] = df['jobinvolvement'].map(sat_map)
```

```python
# Cek hasil mapping satisfaction & involvement
check = ['jobsat_label', 'envsat_label', 'relsat_label', 'jobinv_label']
df[check].head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Mengecek distribusi variabel yang ditarget
df['attrition'].value_counts(normalize=True)
```

```python
# Ubah kolom target 'Attrition' menjadi angka (0 dan 1) utk korelasi
df['attrition_num'] = df['attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
# Mengecek distribusi variabel target (Attrition)
plt.figure(figsize=(6, 4))
sns.countplot(x='attrition', data=df)
plt.title('Distribution of Employee Attrition')
plt.show()
```

```python
# Mengecek hubungan variabel lain vs Attrition (Bivariate Analysis)
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Distribusi Umur berdasarkan Attrition (KDE Plot)
sns.kdeplot(data=df, x='age', hue='attrition', fill=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribusi Umur Karyawan Keluar vs Bertahan')

# Pendapatan vs Attrition (Boxplot)
sns.boxplot(x='attrition', y='monthlyincome', data=df, ax=axes[0,1])
axes[0, 1].set_title('Hubungan Gaji Bulanan dengan Attrition')

# Attrition berdasarkan Job Role (Countplot)
sns.countplot(y='jobrole', hue='attrition', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Jumlah Attrition per Posisi Pekerjaan')

# Korelasi Matriks (Heatmap)
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
corr = df[numeric_cols].corr()
sns.heatmap(corr[['attrition_num']].sort_values(by='attrition_num', ascending=False),
           annot=True, cmap='coolwarm', ax=axes[1, 1])
axes[1, 1].set_title('Korelasi Variabel Numerik terhadap Attrition')

plt.tight_layout()
plt.show()
```
